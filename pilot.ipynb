Define the Persona and Goal

Act as a brand strategist and web consultant. I am building a new personal portfolio website for an interaction and UX designer named Anoushka Rathod. She is a BDes Interaction Design student who works on UX case studies, research-heavy projects, and thoughtful product concepts.

The primary goal of the website is to showcase detailed design case studies and experiments in a way that feels clear, confident, and a little playful. Recruiters, mentors, and collaborators should quickly understand how she thinks, not just how the UI looks.

The brand vibe should be:

Playful but mature, like a young designer who is serious about craft but still curious and experimental

Calm, sky-inspired visuals with a seamless transition from day to night as you scroll down the home page

Interaction-focused, with small delightful details such as animated cursors floating in the hero section to suggest collaboration and “design in progress”

Voice and personality:

“I design thoughtful digital experiences, tell stories with my research, and love turning messy problems into clear, human-friendly products.”

The site should feel opinionated and handcrafted, not like a generic template. Typography and layout should let the work breathe, with strong focus on case study storytelling and process.

Create the Site Map

Home
A strong landing page with the animated sky background (day at the top, night towards the bottom), a bold hero statement, moving cursor arrows, and quick entry points to 3–4 flagship case studies. This page should also include a short “Currently” or “What I am working on” strip and links to GitHub / Dribbble / Behance.

Work
The core of the site. A grid or list of selected projects (case studies). Each project card should show title, role, year, and 1-line problem statement. Clicking a project opens a dedicated case study page.

Case Study Template (project detail page)
A reusable layout used by all projects. Sections: overview, context, problem, research, insights, design process, final solution, and reflection. Include space for images, flows, and quotes. This can be implemented as a dynamic route in FastAPI, pulling content from simple JSON or markdown files.

Experience
A timeline style page showing internships, courses, competitions, and notable collaborations. Each entry should show year, place, role, and one key takeaway. Visually this can sit on a subtle night-sky background to match the “past adventures” feel.

Playground / Experiments
A lighter page for small UI explorations, motion studies, code sketches, and visual experiments that do not need full case studies. Think of it as a gallery of tiles.

About
A personal introduction: who Anoushka is, what she cares about in design, tools she uses, and a short story about her journey so far. Include a friendly photo or illustration.

Contact
A simple page with a contact form (name, email, message) plus direct links to email, GitHub, LinkedIn, and any other relevant profiles. Keep it very easy for a recruiter to reach out.

Home Page Interaction & Visual Direction

Describe this clearly so the implementation follows it:

Full-screen hero section with:

Background: painterly clouds and blue sky artwork

Large serif headline introducing Anoushka and what she does

Supporting line explaining that the site focuses on UX case studies and research-led design

Floating, animated “cursor” arrows in different colours, slowly drifting around the hero text, one with a tiny label like “Can you make it pop” to reference design feedback culture

As the user scrolls:

The day-sky layer gradually fades into a darker, starry night-sky layer using a scroll-based opacity transition, not a hard cut

Content sections (featured work, experience snippet, future interests) scroll over this background with semi-transparent white cards and soft shadows for readability

The navigation bar:

A pill-shaped, slightly glassy bar pinned near the top

Left: “Anoushka Rathod”

Center / right: links for Work, Experience, About, Contact, plus a small button or pill for downloading a Resume

On click, nav items smoothly scroll to that section of the page

Technical Requirements

Use FastAPI as the backend framework.

Serve the portfolio as a mostly static site with:

Jinja2 templates for pages

A static folder for CSS, JavaScript, images, and font files

Make the site responsive for desktop first, but ensure it works well on tablet and mobile.

Use simple, clean CSS (you may use Tailwind or a small custom CSS file) to implement:

The sky background layers and scroll-based opacity transition

The floating cursor animations on the hero

Card-style sections for Work, Experience, and other pages

The final result should be a clean, performant portfolio that can be deployed easily using GitHub Pages + GitHub Actions or another free hosting option available under GitHub Education.
